```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Exploratory Analysis



```python
# column names from covtype.info file
names = [
    "Elevation",
    "Aspect",
    "Slope",
    "Horizontal_Distance_To_Hydrology",
    "Vertical_Distance_To_Hydrology",
    "Horizontal_Distance_To_Roadways",
    "Hillshade_9am",
    "Hillshade_Noon",
    "Hillshade_3pm",
    "Horizontal_Distance_To_Fire_Points",
    "Rawah",
    "Neota",
    "Comanche_Peak",
    "Cache_la_Poudre",
    "Soil_Type_1",
    "Soil_Type_2",
    "Soil_Type_3",
    "Soil_Type_4",
    "Soil_Type_5",
    "Soil_Type_6",
    "Soil_Type_7",
    "Soil_Type_8",
    "Soil_Type_9",
    "Soil_Type_10",
    "Soil_Type_11",
    "Soil_Type_12",
    "Soil_Type_13",
    "Soil_Type_14",
    "Soil_Type_15",
    "Soil_Type_16",
    "Soil_Type_17",
    "Soil_Type_18",
    "Soil_Type_19",
    "Soil_Type_20",
    "Soil_Type_21",
    "Soil_Type_22",
    "Soil_Type_23",
    "Soil_Type_24",
    "Soil_Type_25",
    "Soil_Type_26",
    "Soil_Type_27",
    "Soil_Type_28",
    "Soil_Type_29",
    "Soil_Type_30",
    "Soil_Type_31",
    "Soil_Type_32",
    "Soil_Type_33",
    "Soil_Type_34",
    "Soil_Type_35",
    "Soil_Type_36",
    "Soil_Type_37",
    "Soil_Type_38",
    "Soil_Type_39",
    "Soil_Type_40",
    "Cover_Type",
]

cover_types = pd.DataFrame(
    [
        (1, "Spruce/Fir"),
        (2, "Lodgepole Pine"),
        (3, "Ponderosa Pine"),
        (4, "Cottonwood/Willow"),
        (5, "Aspen"),
        (6, "Douglas-fir"),
        (7, "Krummholz"),
    ],
    columns=["Cover_Type", "covtype_name"],
)

# import data into dataframe
df = (
    pd.read_csv("../data/raw/covtype.data", names=names)
    .merge(cover_types, on="Cover_Type", how="left")
    .drop("Cover_Type", axis=1)
)

wilderness_area = pd.from_dummies(df.iloc[:, 10:14])
soil_type = pd.from_dummies(df.iloc[:, 14:-1])

# add wilderness area
df = pd.concat(
    [df.drop(df.columns[10:-1], axis=1), wilderness_area], axis=1
).rename(columns={"": "Wilderness_Area"})

# add soil type
df = pd.concat([df, soil_type], axis=1).rename(columns={"": "Soil_Type"})
```

```python
# save pairplot of untransformed variables
ax = sns.pairplot(df.sample(1000), hue="covtype_name", corner=True)
plt.savefig("../outputs/untransformed_pairplot.png")
plt.close()
```

```python
df["cos_aspect"] = np.cos(df["Aspect"] * 2 * np.pi / 360)
df["Slope:cos_aspect"] = df["Slope"] * df["cos_aspect"]
df["Hillshade_9am:Hillshade_Noon"] = (
    df["Hillshade_9am"] * df["Hillshade_Noon"]
)
df["Hillshade_9am:Hillshade_3pm"] = df["Hillshade_9am"] * df["Hillshade_3pm"]
df["Hillshade_Noon:Hillshade_3pm"] = (
    df["Hillshade_Noon"] * df["Hillshade_3pm"]
)
df = df.drop("Aspect", axis=1)
```

```python
# save pairplot of untransformed variables
ax = sns.pairplot(df.sample(1000), hue="covtype_name", corner=True)
plt.savefig("../outputs/transformed_pairplot.png")
plt.close()
```

```python
ax = sns.countplot(
    df, 
    x="covtype_name", 
    order=[
        "Spruce/Fir", 
        "Lodgepole Pine", 
        "Ponderosa Pine", 
        "Cottonwood/Willow", 
        "Aspen", 
        "Douglas-fir", 
        "Krummholz"
    ],
    alpha=0.75
)
ax.set_xticklabels(
    labels = [
        "Spruce/Fir", 
        "Lodgepole Pine", 
        "Ponderosa Pine", 
        "Cottonwood/Willow", 
        "Aspen", 
        "Douglas-fir", 
        "Krummholz"
    ],
    rotation = 45,
    ha="right")
plt.tight_layout()
plt.savefig("../outputs/covtype_counts.png")
plt.close()
```

```python
df = (pd.read_csv("../outputs/pca_explained_var/pca_explained_var.csv"))
df["cum_pct_var"] = df.pct_var.cumsum()
```

```python
ax1 = sns.set_style(style=None, rc=None )

fig, ax1 = plt.subplots(figsize=(12,6))

sns.lineplot(data = df['pct_var'], marker='o', sort = False, ax=ax1)
ax2 = ax1.twinx()

sns.barplot(data = df, x='PC', y='cum_pct_var', alpha=0.5, ax=ax2, color="grey")
plt.xlabel('PC')
ax1.set_ylabel('Variance Explained')
ax2.set_ylabel('Variance Explained (Cumulative)')
plt.savefig("../outputs/pca_explained_var.png")
plt.close()
```
